### 0) Libraries

```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

### 1) Path 

#### image extraction from selected path

```python
female_image_path = glob("../../Data/Original_Faces/female/*.jpg") 
male_image_path = glob("../../Data/Original_Faces/male/*.jpg")
```

```python
female_image_path
```

### 2) Data analysis

```python
print(f"Number of female image: {len(female_image_path)}")
print(f"Number of male image: {len(male_image_path)}")
```

### 3) Read image and convert to RGB

#### for 1 image

```python
image = cv2.imread(female_image_path[0])             # image read
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)   # convert image from BRG to RGB
```

```python
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

### 4) Apply Haar Cascade Classifier

```python
haar = cv2.CascadeClassifier("../Trained models/haarcascade_frontalface_default.xml")
gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
faces_list = haar.detectMultiScale(gray, 1.5, 5)
print(faces_list)
for x,y,w,h in faces_list:
    cv2.rectangle(image_rgb, (x,y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
